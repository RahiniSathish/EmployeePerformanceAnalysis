```python
data.corr()
```

```python
# HeatMapAnalysis
plt.figure(figsize=(30,25))
heat=sns.heatmap(data.drop("Performance_Rating",axis=1).corr(),cmap=plt.cm.Accent,annot=True)
plt.title("HeatMapAnalysis")
heat.set_xticklabels(heat.get_xticklabels(), fontsize=12)
heat.set_yticklabels(heat.get_yticklabels(), fontsize=12)
plt.xlabel("columns",fontsize=15)
plt.ylabel("columns",fontsize=15)
plt.show()
```

```python
def correlation(dataset,threshold):
  col_corr=set()
  matrix=dataset.corr()
  for i in range(len(matrix.columns)):
    for j in range(i):
      if(matrix.iloc[i,j]>threshold):
        col_name=matrix.columns[i]
        col_corr.add(col_name)
      return col_corr
```

```python
features=correlation(data,0.4)
len(set(features))
```
