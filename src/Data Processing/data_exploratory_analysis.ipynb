##**`2. Bivariate Analysis`**

###**For Categorical Values**

```python
plt.figure(figsize=(35,30),canvas="white")
plotnumber=1
for column in cat_data:
  if plotnumber<=9:
    ax=plt.subplot(3,3,plotnumber)
    sns.countplot(x=data[column])
    plt.xlabel(column,fontsize=25)
    plt.ylabel("PerformanceRating",fontsize=25)
    plt.xticks(fontsize=20,rotation=45)
  plotnumber+=1
plt.tight_layout()
```

###**For Continuos Values**

```python
plt.figure(figsize=(35,30),canvas="white")  #canvas
plotnumber=1
for column in num_data:
  if plotnumber<=10:
    ax=plt.subplot(3,4,plotnumber)
    sns.histplot(x=data[column],kde=True)
    plt.xlabel(column,fontsize=25)
    plt.ylabel("PerformanceRating",fontsize=25)
    plt.xticks(fontsize=20)
  plotnumber+=1
plt.tight_layout()
```

```python
# Loop through each column for analysis
for column in data:
    # Scatter plot
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=data,x=data[column],y=data['PerformanceRating'])
    plt.title(f'Bivariate Analysis: {column} vs PerformanceRating')
    plt.xlabel(column)
    plt.ylabel('PerformanceRating')
    plt.tight_layout()
    plt.show()
```
