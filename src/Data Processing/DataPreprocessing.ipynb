##**2. Check the Outliers**

```python
#Using Box Plot
plt.figure(figsize=(15,10),canvas="white")
plotnumber=1
for column in num_data:
  if plotnumber<=10:
    ax=plt.subplot(4,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.xlabel(column,fontsize=10)
    plt.ylabel("PerformanceRating",fontsize=10)
  plotnumber+=1
plt.tight_layout()
```

```python
#Handle Outliers for NumCompaniesWorked
Q3=data["NumCompaniesWorked"].quantile(0.75)
print("Upper Quantile",Q3)
Q1=data["NumCompaniesWorked"].quantile(0.25)
print("Lower Quantile",Q1)
```

```python
#Inter Quartile Range(IQR)
IQR=Q3-Q1
IQR
```

```python
#Lower Limit and Upper Limit
Lower_Limit=Q1-1.5*IQR
print("Lower_Limit:",Lower_Limit)
Upper_Limit=Q3+1.5*IQR
print("Upper_Limit:",Upper_Limit)
```

```python
data.loc[data["NumCompaniesWorked"]<Lower_Limit]
```

```python
data.loc[data["NumCompaniesWorked"]>Upper_Limit]
```

```python
len(data.loc[data["NumCompaniesWorked"]>Upper_Limit])/len(data)
```

###The Value is less than 5% So, we replace with Median

```python
data.loc[data["NumCompaniesWorked"]>Upper_Limit,"NumCompaniesWorked"]=data["NumCompaniesWorked"].median()
```

```python
sns.boxplot(x=data.NumCompaniesWorked)
```

```python
#Using Box Plot
plt.figure(figsize=(15,10),canvas="white")
plotnumber=1
for column in num_data:
  if plotnumber<=10:
    ax=plt.subplot(4,3,plotnumber)
    sns.boxplot(x=data[column])
    plt.xlabel(column,fontsize=10)
    plt.ylabel("Attrition",fontsize=10)
  plotnumber+=1
plt.tight_layout()
```

###Here, Successfully Removed the Outliers.
